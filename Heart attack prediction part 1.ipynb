```python
columns_list = ['PhysicalHealthDays', 'MentalHealthDays', 'SleepHours','HeightInMeters', 'WeightInKilograms', 'BMI']
print('the shape before removing outliers-->',df.shape[0])



def remove_outliers_iqr(df, columns_list):
    """
    Remove outliers from specific columns of a pandas DataFrame using the IQR method.

    Parameters:
    df (pd.DataFrame): The DataFrame from which to remove outliers.
    columns_list (list): The list of columns for which to remove outliers.

    Returns:
    pd.DataFrame: The DataFrame with outliers removed from the specified columns.
    """
    # For each column in columns_list, calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df[columns_list].quantile(0.25)
    Q3 = df[columns_list].quantile(0.75)

    # Calculate the Interquartile Range (IQR)
    IQR = Q3 - Q1

    # Define lower and upper bounds
    lower_bound = Q1 - 3.25 * IQR
    upper_bound = Q3 + 3.25 * IQR

    # Remove rows with outliers in the specified columns
    df_cleaned = df[~((df[columns_list] < lower_bound) | (df[columns_list] > upper_bound)).any(axis=1)]

    return df_cleaned

# Apply the function to remove outliers from the specified columns
df_cleaned = remove_outliers_iqr(df, columns_list)

# Print the shape of the cleaned DataFrame
print('the shape after removing outliers-->',df_cleaned.shape[0])
```
